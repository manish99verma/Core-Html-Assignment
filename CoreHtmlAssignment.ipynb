```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Document</title>
  </head>
  <body>
    <ol>
      <li>
        <strong>&lt;html&gt;</strong> - The root element of an HTML document
        that contains all other elements.
      </li>
      <li>
        <strong>&lt;head&gt;</strong> - Contains metadata such as the title,
        character encoding, and linked stylesheets.
      </li>
      <li>
        <strong>&lt;title&gt;</strong> - Sets the title of the web page, which
        appears in the browser tab.
      </li>
      <li>
        <strong>&lt;body&gt;</strong> - Holds all the visible content of the web
        page.
      </li>
      <li>
        <strong>&lt;h1&gt; to &lt;h6&gt;</strong> - Define headings, where
        &lt;h1&gt; is the largest and &lt;h6&gt; is the smallest.
      </li>
      <li><strong>&lt;p&gt;</strong> - Represents a paragraph of text.</li>
      <li>
        <strong>&lt;a&gt;</strong> - Defines a hyperlink, allowing users to
        navigate to another page or section.
      </li>
      <li>
        <strong>&lt;img&gt;</strong> - Embeds an image using the "src" attribute
        to specify the image URL.
      </li>
      <li>
        <strong>&lt;ul&gt;</strong> - Creates an unordered list, typically
        displayed with bullet points.
      </li>
      <li>
        <strong>&lt;ol&gt;</strong> - Creates an ordered list, typically
        displayed with numbers.
      </li>
      <li>
        <strong>&lt;li&gt;</strong> - Represents a list item inside &lt;ul&gt;
        or &lt;ol&gt;.
      </li>
      <li>
        <strong>&lt;div&gt;</strong> - A generic container used to group and
        style elements.
      </li>
      <li>
        <strong>&lt;span&gt;</strong> - A small inline container used for
        styling a part of text.
      </li>
      <li><strong>&lt;table&gt;</strong> - Creates a table structure.</li>
      <li><strong>&lt;tr&gt;</strong> - Represents a row in a table.</li>
      <li>
        <strong>&lt;td&gt;</strong> - Represents a data cell in a table row.
      </li>
      <li>
        <strong>&lt;th&gt;</strong> - Represents a header cell in a table row.
      </li>
      <li>
        <strong>&lt;form&gt;</strong> - Defines an interactive form for user
        input.
      </li>
      <li>
        <strong>&lt;input&gt;</strong> - Defines an input field in a form (e.g.,
        text, checkbox, radio).
      </li>
      <li><strong>&lt;button&gt;</strong> - Represents a clickable button.</li>
    </ol>
  </body>
</html>
```
